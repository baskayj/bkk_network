```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('route_deg_corr_nighttime.csv', header = 0)
```

```python
neutral = 20.1927710843373494

fig = plt.figure()
plt.plot(df['k'],df['knn'], 'o')
plt.hlines(neutral,0,30, color = 'black')

fig.patch.set_facecolor('white')
plt.legend(('Measured','$\\frac{<k^2>}{<k>}=$'+f'{round(neutral,2)}'))
plt.xlabel('k')
plt.ylabel("$k_{nn}(k)$")
plt.grid()
plt.title('Degree Correlation Function of nighttime BpPT Routes Network')
plt.savefig('route_deg_corr_nighttime.png')
```
