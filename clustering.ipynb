```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# ROUTES - NIGHTTIME
df = pd.read_csv('route_clustering_nighttime.csv',header=0)
```

```python
mu = np.mean(df['clustering'])
mu_global = 0.652

fig = plt.figure()
hist,bins, patches = plt.hist(df['clustering'],bins = 5, density=True)
plt.vlines(mu,0,3.9, colors="black")
plt.vlines(mu_global,0,3.9, colors="red")


fig.patch.set_facecolor('white')
plt.legend((f'$<C_i>$ = {round(mu,2)}',f'C = {round(mu_global,2)}','Distribution'))
plt.xlabel('$C_i$')
plt.ylabel('$p(C_i)$')
plt.grid()
plt.title('Distribution of the Clustering Coefficient in the nighttime BpPT Routes Network')
plt.savefig('route_clustering_nighttime.png')
```
